```python
#VKİ Hesaplama
df['BodyMassIndex'] =  df['MaxBodyMass']/((df['MaxLength']*df['MaxLength'])*10)

plt.figure(figsize=(10, 6))
plt.scatter(df['MaxBodyMass'], df['MaxLength'], c=df['BodyMassIndex'], cmap='viridis', s=100, edgecolors='black', linewidths=1, alpha=0.75)


cbar = plt.colorbar()
cbar.set_label('Body Mass Index')

plt.title('MaxWingspan vs MaxBodyMass Scatter Plot')
plt.xlabel('Max BodyMass')
plt.ylabel('Max Length')
plt.grid(True)
plt.show()
```
