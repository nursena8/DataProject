```python
def make_frame(country:str):
  df=pd.DataFrame({
      "infected":df_infected.loc[country],
      "deaths":df_deths.loc[country],
      "recoverd":df_recoverd.loc[country]
  })
  df.index=pd.to_datetime(df.index,format='%m/%d/%y')
  return df
```

```python
make_frame("China")
df_turkey=make_frame("Turkey")
```
