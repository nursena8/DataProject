```python
job_listings = []


for job_elem in text_soup.find_all('div', class_='media-content'):
    title_elem = job_elem.find('h2', class_='title is-5')

    if None in (title_elem):
        continue

    job_listings.append({
        'title': title_elem.get_text(strip=True),
    })

#  ilanlarn grselletir
    # print(job_listings)
for job in job_listings:
    print(f"Balk: {job['title']}")
    print("-" * 30)
```

```python
job_elements=text_soup.find_all("div",class_="card-content")
```

```python
for job_element in job_elements:
   title_element=job_element.find("h2",class_="title")
   print(title_element)
   company_element=job_element.find("h3",class_="company")
   location_element=job_element.find("p",class_="location")
   print(location_element,company_element)
```

```python
print(f"there are total {len(job_elements)} jobs")
for num,element in enumerate(job_elements):
  title_element=element.find("h2",class_="title")
  company_element=element.find("h3",class_="company")
  location_element=element.find("p",class_="location")
  print(num+1)
  print(title_element.text.strip())
  print(company_element.text.strip())
  print(location_element.text.strip())
  print("\n")
```

```python
job_listings = []

for job_elem in text_soup.find_all('div', class_='media-content'):
    title_elem = job_elem.find('h2', class_='title is-5')

    if None in (title_elem):
        continue

    title_text = title_elem.get_text(strip=True)

    if 'Python' in title_text:
        job_listings.append({'title': title_text})

for job in job_listings:
    print(f"Balk: {job['title']}")
    print("-" * 30)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
from collections import Counter

all_job_titles = " ".join(job['title'] for job in job_listings)
word_freq = Counter(all_job_titles.split())


res = word_freq.most_common(20)

wc = WordCloud(background_color='white', width=800, height=600)

plt.figure(figsize=(15, 7))
plt.imshow(wc.generate_from_frequencies({k: v for k, v in res}))
plt.axis("off")
plt.show()
```
