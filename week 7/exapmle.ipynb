```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-5.0,5.0,0.1)
y= 3 + 2*(x)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
```

```python
plt.figure(figsize=(12,8))
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
```

```python
x = np.arange(-5.0,5.0,0.1)

y= 3 + x**3 + x**2 + 1*x
y_noise = 20 * np.random.normal(size=x.size)
ydata=y + y_noise
```

```python
plt.figure(figsize=(12,8))
plt.plot(x , ydata,"bo")
plt.plot(x,y,"r")
plt.xlabel("dependent variable")
plt.ylabel("indipendent varibale")
```

```python
x_data,y_data=(df["Year"].values,df["Value"].values)
```

```python
plt.plot(x_data,y_data,"ro")
```

```python
np.log(x_data) # logaritmic
np.polyfit(np.log(x_data),y_data,1)
lo_fit = np.polyfit(np.log(x_data), y_data, 1)
log= np.poly1d(lo_fit)
y_fit = log(np.log(x_data))
plt.scatter(x_data, y_data, label='Orijinal Veri', color='blue')
plt.plot(x_data, y_fit, label='Uyumlanm Eri', color='red')
```

```python
quad_fit = np.polyfit(x_data, y_data, 2)
quad_eq = np.poly1d(quad_fit)
y_fit = quad_eq(x_data)

plt.scatter(x_data, y_data, label='Orijinal Veri', color='blue')
plt.plot(x_data, y_fit, label='Uyumlanm Eri', color='red')
```

```python
exp_fit = np.polyfit(x_data, np.log(y_data), 1)
exp_eq = np.poly1d(exp_fit)
y_fit = np.exp(exp_eq(x_data))
plt.scatter(x_data, y_data, label='Orijinal Veri', color='blue')
plt.plot(x_data, y_fit, label='Uyumlanm Eri', color='red')
```
