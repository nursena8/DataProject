```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches
import seaborn as sns
import numpy as np
```

```python
df=pd.DataFrame({
    "Group":["a","b","c","d","e"],"value":[1,3,5,9,13]})
```

```python
my_plot=plt.barh(y=df.Group,width=df.value,color=mcolors.TABLEAU_COLORS,edgecolor="black",linewidth=3)
plt.bar_label(container=my_plot,padding=4)
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
tips.info()
```

```python
sns.set(style="darkgrid")
```

```python
sns.barplot(x="total_bill",y="day",data=tips,color="cyan")
plt.figure(figsize=(18,8))
```

```python
g=sns.FacetGrid(tips,col="day",height=4,aspect=.5)
g.map(sns.barplot,"sex","total_bill")
```

```python
sns.barplot(x="total_bill",y="sex",hue="time",data=tips,palette="dark:pink")
plt.figure(figsize=(18,6))
```

```python
total=tips.groupby("day")["total_bill"].sum().reset_index()
```

```python
total
```

```python
bar_1=sns.barplot(x="day", y="total_bill", data=total)
```

```python
filt=tips[tips["sex"]=="Female"]
```

```python
bar2=sns.barplot(x="day",y="total_bill",data=filt,estimator=sum,errorbar=None,color="lightblue")
```

```python
bar1=sns.barplot(x="day",y="total_bill",data=total,color="blue")
bar2=sns.barplot(x="day",y="total_bill",data=filt,estimator=sum,errorbar=None,color="lightblue")
top_bar=mpatches.Patch(color="darkblue",label="not female")
bottom_bar=(mpatches.Patch(color="lightblue",label="female"))
plt.legend(handles=[top_bar,bottom_bar])
```

```python
df_iris=sns.load_dataset("iris")
```

```python
sns.histplot(x="sepal_length",data=df_iris,kde=True)
```

```python
sns.regplot(x=df_iris["sepal_length"],y=df_iris["sepal_width"],marker="+",line_kws={
    "color":"red",
    "alpha":0.6,
    "lw":5
},scatter_kws={"color":"darkblue","s":50})
```

```python
sns.lmplot(data=df_iris,x="sepal_length",y="sepal_width",hue="species")
```

```python
sns.set_theme(style="white",palette="pastel")
sns.boxplot(x=df_iris["species"],y=df_iris["sepal_length"])
```

```python
df_iris.boxplot(column="sepal_length",by="species")
```
